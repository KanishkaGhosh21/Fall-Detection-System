```python
import pandas as pd
train = pd.read_csv('raw182_Training_Relabeled_Auto_25.csv')
train.head()
```

```python
X = train.drop(['outcome'], axis=1)
y = train['outcome']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
rf_random.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(
n_estimators = 400,
min_samples_split = 5,
 min_samples_leaf =1,
 max_features ='auto',
 max_depth =30,
 bootstrap =False
)
clf = clf.fit(X, y)
```

```python
X_t = train.drop(['outcome'], axis=1)
y_t = train['outcome']
X_t = sc.fit_transform(X)
```

```python
# Create our predictions
prediction = clf.predict(X_t)
# print all predictions that are 1
for i in range(len(prediction)):
    if prediction[i] == 1:
        # print with the X_test index
        print(i, prediction[i], X_t[i])
```

```python
y_t
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
accuracy_score(y_t, prediction)
```

```python
cm = confusion_matrix(y_t, prediction)
import matplotlib.pyplot as plt
import seaborn as sns

# Change figure size and increase dpi for better resolution
plt.figure(figsize=(8,6), dpi=100)
# Scale up the size of all text
sns.set(font_scale = 1.1)

# Plot Confusion Matrix using Seaborn heatmap()
# Parameters:
# first param - confusion matrix in array format   
# annot = True: show the numbers in each heatmap cell
# fmt = 'd': show numbers as integers. 
ax = sns.heatmap(cm, annot=True, fmt='d', )

# set x-axis label and ticks. 
ax.set_xlabel("Predicted", fontsize=14, labelpad=20)
ax.xaxis.set_ticklabels(['No', 'Yes'])

# set y-axis label and ticks
ax.set_ylabel("Fall Detection", fontsize=14, labelpad=20)
ax.yaxis.set_ticklabels(['No', 'Yes'])

# set plot title
ax.set_title("Confusion Matrixl", fontsize=14, pad=20)

plt.show()
```
